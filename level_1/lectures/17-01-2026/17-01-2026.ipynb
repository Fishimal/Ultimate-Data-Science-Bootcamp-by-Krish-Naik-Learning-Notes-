### Naming Convention
- variable name can contain letters, digits or underscore (_)
    - bmi, person0, person_1, _person, id_1
- it should not start with a digit
- variable names are case-sensitive: hello, Hello, heLlo
- it should not be a Python keyword: print, type, id, if, else, for

### Flow of Python
- In Python execution happens from top to bottom and left to right
- In case of parenthesis it starts from the inner bracket and the last one is the outer bracket

```python
mentor_name = "Monal Singh"
print(mentor_name)
value_1 = 100
print(500)
value_2 = 200
print(value_1)
```
