## 1. The Office Door Badge 
You are programming a smart door for a tech office. The door should only open if the user provides the correct "Access Code."
- **Input:** User enters a 4-digit code.
- **Scenario:** If the code is 1234, print "Access Granted. Door Unlocked." Otherwise, print "Access Denied. Security Alerted."

```python
access_code = int(input("Please enter the 4-digit access code to open the door : ").strip())

if access_code == 1234:
    print("Access Granted. Door Unlocked")
else:
    print("Access Denied. Security Alerted")
```

## 2. The Weekend Grocery Trip 
A person is deciding whether to go to the supermarket.
- **Input:** User enters how many eggs are left in the fridge (integer).

## 3. The Smart Home "Leaving for Work" Checklist 
When you leave your house, a smart assistant checks multiple independent things. These are not mutually exclusive.
- **Input:** Three separate inputs (Yes/No) for: "Is the AC on?", "Is the Coffee Maker on?", "Is the Front Door
unlocked?"
- **Scenario:**
    - If AC is on Print "Turning off AC to save energy."
    - If Coffee Maker is on Print "Warning: Coffee Maker is still hot!"
    - If Door is unlocked Print "Locking the front door now." (Note: All three messages could potentially print.)

```python
is_AC_on = input("Is the AC on? ").strip().capitalize()
is_coffe_maker_on = input("Is the Coffee Maker on? ").strip().capitalize()
is_front_door_unlocked = input("Is the Front Door unlocked? ").strip().capitalize()

if is_AC_on == "Yes":
    print("Turning off AC to save energy.")

if is_coffe_maker_on == "Yes":
    print("Warning: Coffee Maker is still hot!")

if is_front_door_unlocked == "Yes":
    print("Locking the front door now.")
```

## 4. The Pizza Topping Calculator 
A customer is choosing toppings. Each topping adds a specific cost regardless of the other toppings chosen.
- **Input:** Three boolean (True/False) inputs for: <mark>extra_cheese</mark>, <mark>pepperoni</mark>, and <mark>olives</mark>.
- **Scenario:** Initialize a <mark>total_bill = 10</mark>.

    - If <mark>extra_cheese</mark> is True add 2 to bill.

    - If <mark>pepperoni</mark> is True add 3 to bill.

    - If <mark>olives</mark> is True add 1 to bill.
- **Output:** Print the final total bill.

```python
total_bill = 10

print("Customize your pizza with additional toppings.")

extra_cheese = input("Do you want extra_cheese: ").strip()
pepperoni = input("Do you want to add pepperoni: ").strip()
olives = input("Do you want to add olives: ").strip()

if extra_cheese.capitalize() == "True":
    total_bill += 2

if pepperoni.capitalize() == "True":
    total_bill += 3

if olives.capitalize() == "True":
    total_bill += 1

print(f"Your total bill = {total_bill}")
```

## 5. The Traffic Light Sensor 
A self-driving car needs to interpret a single traffic light color.
- **Input:** User enters "Red", "Yellow", or "Green".
- **Scenario:** 
    - If "Red" "Full Stop."
    - If "Yellow" "Slow down and prepare to stop."
    - If "Green" "Proceed with caution."
    - Anything else "Error: Signal Malfunction!"

```python
signal_traffic_light_color = input("Enter the traffic light color: ").strip()

if signal_traffic_light_color.capitalize() == "Red":
    print("Full Stop.")
elif signal_traffic_light_color.capitalize() == "Yellow":
    print("Slow down and prepare to stop.")
elif signal_traffic_light_color.capitalize() == "Green":
    print("Proceed with caution.")
else:
    print("Error: Signal Malfunction!")
```

## 6. Movie Theater Ticket Pricing 
A theater charges based on the age of the customer.
- **Input:** User enters their age.
- **Scenario:**
    - Age < 5 "Ticket is Free."
    - Age 5 to 17 "Child Ticket: $10."
    - Age 18 to 60 "Adult Ticket: $18."
    - Age > 60 "Senior Ticket: $12."

```python
age = int(input("Enter your age to get your ticket charges: ").strip())

if (age < 5):
    print("Ticket is Free.")
elif (5 <= age <= 17):
    print("Child Ticket: $10.")
elif (18 <= age <= 60):
    print("Adult Ticket: $18.")
else:
    print("Senior Ticket: $12.")
```

## 7. The 7-Day Steps Tracker 
A fitness app wants to show your activity for the week.
- **Data:** A list of steps: [4000, 7500, 10000, 2000, 8000, 12000, 5000].
- **Scenario:** Loop through the list and for each day, print: "Day [X]: You walked [Steps] steps."

```python
# Solution - 1
list_of_steps = [4000, 7500, 10000, 2000, 8000, 12000, 5000]
day = 1

for step_count in list_of_steps:
    print(f"Day {day}: You walked {step_count} steps.")
    day += 1
```

```python
# Solution - 2
list_of_steps = [4000, 7500, 10000, 2000, 8000, 12000, 5000]

for day, step_count in enumerate(list_of_steps, start=1):
    print(f"Day {day}: You walked {step_count} steps.")
```
